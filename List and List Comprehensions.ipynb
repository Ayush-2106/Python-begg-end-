## List Comprehension
- Basics Syntax{Expression for item in iterable}
- With conditional Logic{expression for item in iterable if condition}
- Nested list comprehension {expression for item in iterable1} for item2 in iterable2

```python
lst =  []

for i in range(1, 11):
  lst.append(i)
print(lst)
```

```python
[i for i in range(1, 11)]
```

```python
lst =  []

for i in range(1, 11):
  lst.append(i*i)
print(lst)
```

```python
[i*i for i in range(1, 11)]
```

```python
lst =  []

for i in range(1, 51):
  if i % 2 == 0:
    lst.append(i)
print(lst)
```

```python
print([i for i in range(1, 51) if i % 2 == 0])
```

```python
lst = []

for i in range(1, 7):
  for j in range(1, 7):
    lst.append((i,j))

print(lst)
```

```python
print([(i,j) for i in range(1, 7) for j in range(1, 7)])
```

```python
print([(i,j) for i in range(10,20) for j in range(10,20)])
```
